```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
```

```python
df = pd.read_csv('Diwali Sales Data.csv' , encoding= 'unicode_escape')
```

```python
df.dropna(inplace = True)
```

## EDA 

```python
## Gender 

sns.countplot(x = 'Gender' , data=df)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)

sns.barplot(x='Gender' , y= 'Amount' , data=sales_gen)
```

## Age

```python
Age_Group_plot = sns.countplot(data=df,x='Age Group' , hue='Gender')

for bars in Age_Group_plot.containers:
    
    Age_Group_plot.bar_label(bars)
```

```python
df.columns
```

```python
sales_age_group = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)

sns.barplot(x='Age Group' , y= 'Amount' , data=sales_age_group)
```

## State

```python
df.columns
```

```python
sales_age_group = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders' , ascending=False)

sns.set(rc={'figure.figsize':(19,6)})
sns.barplot(x='State' , y= 'Orders' , data=sales_age_group)
```

```python
sales_age_group = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)

sns.set(rc={'figure.figsize':(19,6)})
sns.barplot(x='State' , y= 'Amount' , data=sales_age_group)
```

```python
M_S = sns.countplot(data=df , x='Shadi')

#sns.set(rc={'figure.figsize':(2,6)})
plt.figure(figsize=(2, 6))
```
